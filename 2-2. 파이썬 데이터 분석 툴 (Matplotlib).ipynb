## Matplotlib에서 다양한 그래프 나타내기

Matplotlib을 통해 다양한 형태의 그래프도 그릴 수 있습니다.

### Scatter

- scatter(x, y, c, s, alpha...) 
    - c : 색상 설정
    - s : 크기 설정
    - alpha : 투명도 설정  
    
### Bar
- bar(x,y) : x, y 데이터를 바 형태로 출력

### Multi-Bar
- set_xticks(x) : x 축의 데이터를 병렬적으로 설정
- set_xticklabels() : x축 label 설정

### Histogram
- hist(data, bins) 
    - data : 데이터 입력
    - bins : 히스토그램 표현시 분할되는 개수 설정

```python
import numpy as np
import matplotlib.pyplot as plt


def main():
    fig, axes = plt.subplots(2,2)
    
    # Scatter 그래프 그리기
    x = np.random.randn(50)
    y = np.random.randn(50)
    colors = np.random.randint(0,100, 50)
    sizes = 500*np.pi*np.random.rand(50)**2
    axes[0,0].scatter(x, y, c=colors, s=sizes, alpha=0.3)
    
    
    # Bar 그래프 그리기
    x = np.arange(10)
    axes[0,1].bar(x, x**2)
    
    
    # Multi-Bar 그래프 그리기
    x = np.random.rand(3)
    y = np.random.rand(3)
    z = np.random.rand(3)
    data =  [x, y, z]

    x_ax =  np.arange(3)
    for i in x_ax:
        axes[1,0].bar(x_ax, data[i], bottom=np.sum(data[:i], axis=0))
    axes[1,0].set_xticks(x_ax)
    axes[1,0].set_xticklabels(['A', 'B', 'C'])
    
    
    # Histogram 그래프 그리기
    data = np.random.randn(1000)
    axes[1,1].hist(data, bins=50)
    
    fig.savefig("plot.png")
    
if __name__ == "__main__":
    main()
```

### 화려한 그래프를 그리고 싶다면? : Seaborn

Matplotlib를 기반으로 다양한 색상 테마, 통계용 차트 기능이 추가된 패키지

```python
# 데이터 불러오기

import seaborn as sns
from matplotlib import pyplot as plt
df = sns.load_dataset('iris')
df.head()

# 산점도 그리기

sns.regplot(x=df["sepal_length"],y=df["petal_length"],color='red',line_kws={'color':'blue'})
# 선의 속성만 변경 line_kws={'color':'blue'}
# 점의 속성만 변경 scatter_kws={'color':'red'}
```
